# 데이터 전처리

### 데이터-> 정보: 속성이 포함되어있고 잘 정리된 것
#### 데이터 축소: 
* 차원축소;시간이 비약적으로 줄여줌, 차원축소는 필수!(너무 세세한 차이까지 찾아내기때문에, 숲보다 나무를 보는 단점)
* '1년 동안 취미를 즐긴 횟수' 23번, 42번 이런거 삭제, '최근 1년 내 취미활동을 즐겼나요?' 로 1,0 으로 수량 감소
* 데이터 속성 제거: 중복 속성 제거 (최근 일년 간 즐긴 취미, 취미 둘중 하나 제거)
* 샘플링:특정한 집단이 있을때 그 중 하나를 선정 -> 모집단, 데이터가 너무 많으면 추려서 분석, 속도 빨라짐

#### 데이터 이산화
* 연속형 데이터(스칼라): 숫자로 나타낸 값, 쪼갤 수 있는 것, 마땅한 기준x
* 범주형 데이터(카테고리컬): ex)성별
* 데이터 이산화: 연속형 데이터 -> 범주형 데이터 변경, 
* ex) 큰키:170이상, 작은키:160이하
* 데이터 이산화를 하는 이유: 컴퓨터는 판단을 못하기 때문에 직접 기준을 세워줘야함.


#### 표준화, 정규화
* 표준화(standardization): 정규분포의 속성을 갖도록 데이터 분포가 조절됨,
    데이터 범위의 제한 없음, Z-score 등
* 정규화(Normalization):
    * 그라디언트 배니싱: 딥러닝이 계산하는 과정에서 기울기가 0이 되어버리는 상황
    * 데이터의 분포 차이가 유지됨
    * 데이터 범위의 제한이 생김
    * min-max scaling 등
    * -10~999 -> 0~1 로 변환 : 정규화
