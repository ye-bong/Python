4. 선택한 연산에 따라 연산 함수를 실행합니다.
5. 연산 함수의 결과값을 출력합니다.

```python
def cal(x,y):
    a = str(input("어떤 연산을 하시겠어요?(더하기,빼기,곱하기,나누기)"))
    if a == '더하기':
        return x+y
    elif a == '빼기':
        return x-y
    elif a == '곱하기':
        return x*y
    else:
        return x/y
```

```python
cal(2,3)
```

```python
cal(2,3)
```

# 함수실습
* 함수 이름: func1
* 입력받는 값: 임의의 숫자 둘(x,y)
* 반환값:x/y의 나머지와 x-y를 비교해
* 전자가 크면 Type1
* 둘이 같거나 후자가 크면 type2 라고 반환

```python
def func1(x,y):
    if x%y > x-y:
        return 'Type1'
    else:
        return 'Type2'
```

```python
func1(2,3)
```

```python
2%3
```

```python
2-3
```

```python
func1(2,1)
```

### 강사님 코드

```python
def func1(x,y):
    A = x%y
    B = x-y
    
    if A > B:
        return 'Type1'
    elif A <= B:
        return 'Type2'
```

```python
func1(2,1)
```
