```python
# 문제20 Age가 짝수인 경우 1, 홀수인 경우 0으로 반환하는 함수를 작성하라
```

```python
titanic['Age_c'] = titanic['Age']%2==0
```

```python
titanic['Age_c'] = titanic['Age_c'].apply(lambda x: int(x))
```

```python
titanic
```

```python
def age_chk(x):
    if x % 2 == 1:
        return 1
    elif x % 2 == 0:
        return 0
    else:
        return float('nan')
```

```python
for j in [age_chk(i) for i in titanic['Age']]:
    print(j)
```
