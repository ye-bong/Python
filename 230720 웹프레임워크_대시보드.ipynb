```python
from dash import Dash, html, dcc, callback, Output, Input, dash_table # 딕셔너리를 받음
import plotly.express as px
import pandas as pd

df = sns.load_dataset('iris')

# df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder_unfiltered.csv')

app = Dash(__name__)

app.layout = html.Div([
    html.H1(children='iris 산점도 그래프', style={'textAlign':'center'}),
    dcc.Dropdown(df['species'].unique(), 'setosa', id='dropdown-selection'),
    dcc.Graph(id='graph-content'),
    dash_table.DataTable(df.to_dict('records'), [{"name": i, "id": i} for i in df.columns])
])

@callback(
    Output('graph-content', 'figure'),  # id, 형식
    Input('dropdown-selection', 'value')
)
def update_graph(value):
    dff = df[df['species']==value] # 마스킹
    return px.scatter(dff, x='sepal_length', y='sepal_width')

if __name__ == '__main__':
    app.run(debug=True) # debug=True 변경사항이 있을때 즉시 새로고침을 하겠다.
```
