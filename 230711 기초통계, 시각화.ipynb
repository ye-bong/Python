# 기초통계

## 표준화, 정규화
- 표준화: 데이터 집단 표준편차를 1로 제한하는 변형하는 Z-score
- 정규화: 모든값이 0~1 사이로 만드는 것
- 정규분포
    - 특정 값 - 평균(또는 중앙값)
    - 표준편차: 분산^(0.5)
    - 분산

## 가설
- 어떠한 사실을 설명 또는 증명하기 위한 가정
- 귀무가설(영가설, Null Hypothesis, H1):
    - 가설 검정의 대상이 되는 가설
    - 대개 기각될 것으로 예상되는 가설
    - 증명할 수 있는 능력이 큰 것을 가설로 설정 (내 의견과 가장 반대되는 의견)
    - ex) 지구는 둥글지 않다. 왜냐면 어쩍구 저쩌구
- 대립가설(Anti Hypothesis, H0):
    - 가설검정에서 입증하려 하는 가설
- 가설을 채택 또는 기각하여 논리 전개

- 가설 채택 != 해당 가설이 참이다
- 가설 기각 != 해당 가설이 거짓이다
- 가설 채택 = 가설을 기각할 근거가 충분하지 않음
- 가설 기각 = 가설을 기각할 근거가 있음
### 제 1종 오류와 제 2종 오류
- 귀무가설 맞지만 귀무가설 기각한 경우 : 제 1종 오류
- 대립가설 맞지만 귀무가설 기각X: 제 2종 오류
### 유의수준 (p-value, p, 알파)
- 제 1종 오류를 범할 확률
- 귀무가설이 참임에도 불구하고 기각할 확률
- cutoff 수준: 0.05(5%), 0.01, 0.001
### T-검정 (T-test)
- T분포를 이용한 통계적 검정
- 정의: A, B 그룹이 있을때,  통계적으로 차이가 있는 지 검증하는 방법
    - 정규분포를 따르는 모집단 N1 ,N2에서 추출된 표본자료들이 모평균이 같은지 검정
- '두 그룹이 같다 다르다, 분포가 같다 다르다'를 판단
- T-test의 조건:
    - 1. 양적 변수 (셀 수 있는 변수)
    - 2. 정규분포를 따름
    - 3. 등분산성 충족 (분산이 같음)
- T-test의 절차
    - 1. 집단 특성 확인
        - 1. 독립표본 -> 아래 절차로 이동
        - 2. 대응표본(ex: 코딩테스트 전날 치룬것과 오늘 치룬것의 차이 검증) -> Paired T-test
    - 2. 자료의 특성 확인
        - 1. 범주형(~번째, 랭크를 메길 수 있는 경우만) -> Mann-Whitney U test
    - 3. 연속형일 경우
        - 1. 정규성, 등분산성 만족 -> Student's T-test
        - 2. 정규성 만족, 등분산성 불만족 -> Welch's T-test
        - 3. 정규성 불만족 -> Mann-Whitney U test

```python
m1 = df_iris['species'] == 'setosa'
m2 = df_iris['species'] == 'versicolor'
```

```python
from scipy import stats
```

```python
ttest = stats.ttest_ind(setosa, versicolor)
```

```python
# t-test의 귀무가설 : 두 그룹의 값들이 서로 차이가 나지 않는다.
# p가 0.05 보다 작으면 유의한 차이가 있다.
```

```python
sns.boxplot(data=mpg, x='category', y='horsepower')
sns.stripplot(data=mpg, x='category', y='horsepower', color='k', alpha=.5)
```

```python
f1 = sns.boxplot(data=mpg, x='category', y='horsepower')
f2 = sns.stripplot(data=mpg, x='category', y='horsepower', color='k', alpha=.5)

add_stat_annotation(f1, data=mpg, x='category', y='horsepower',
                   box_pairs=[(('over'),('under'))],
                   test='t-test_ind', text_format='star',
                   loc='inside', verbose=2)
```

```python
# user_df: 입력할 데이터프레임 이름
# col: x축 column 이름
# value: y축 value 이름
# group1, group2: col 기준으로 분류된 이름

def ttest_auto(user_df, col, value, group1, group2):
    f1 = sns.boxplot(data=user_df, x=col, y=value)
    f2 = sns.stripplot(data=user_df, x=col, y=value, color='k', alpha=.5)

    add_stat_annotation(f1, data=user_df, x=col, y=value,
                       box_pairs=[((group1),(group2))],
                       test='t-test_ind', text_format='star',
                       loc='inside', verbose=2)
    
    plt.show()
```
